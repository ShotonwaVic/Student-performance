```python
from sklearn.preprocessing import LabelEncoder
```

```python
new_diagnosis = LabelEncoder()
```

```python
data["new_diagnosis"] = new_diagnosis.fit_transform(data["diagnosis"])
```

```python
data
```

```python
data = data.drop("diagnosis", axis = 1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = data.drop(columns = "new_diagnosis", axis = 1)
```

```python
y = data["new_diagnosis"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, classification_report, f1_score, recall_score
```

```python
models = {"Logistics Regression": LogisticRegression(),
        "Decision Tree Classifier": DecisionTreeClassifier(),
        "Random Forest Classifier": RandomForestClassifier(),
        "Support Vector Mechanics": SVC(),
        "K-Neighbour Classifier": KNeighborsClassifier()}
```

```python
results = {}
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print (f"The Prediction is {name}: {y_pred}\n")
    accuracy = accuracy_score(y_test, y_pred)
    #lass_report = classification_report(y_test, y_pred)
    results[name] = {'Accuracy': accuracy, 'Classification Report': classification_report(y_test, y_pred, zero_division= 0)}
```

```python
results
```

# To check result

```python
for name, metrics in results.items():
    print(f"{name}")
    print (f"Accuracy of Model: {metrics['Accuracy'] * 100:.2f}%")
    print (metrics['Classification Report'])
    print("\n")
```
