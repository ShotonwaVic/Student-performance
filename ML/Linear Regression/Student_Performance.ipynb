# Splitting our data into x and y

```python
x = data.drop(columns = "Performance Index")
```

```python
x
```

```python
y = data["Performance Index"]
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Extra_activities = LabelEncoder()
```

```python
x["Extra_activities"] = le_Extra_activities.fit_transform(x["Extracurricular Activities"])
```

```python
x
```

```python
x = x.drop(["Extracurricular Activities"], axis = 1)
```

```python
x
```

# Train - Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
print(f"intercept is {model.intercept_}")
```

The intercept of  (-34.075) In linear regression, the intercept represents the expected value of the dependent variable (y) when all independent variables (x) are equal to zero.

- Interpretation: 
    The intercept of -34.075 means that: "If a student has: 
    - 0 hours studied 
    - 0 previous scores 
    - 0 extracurricular activities 
    - 0 sleep hours 
    - 0 simple question paper practiced 
 
 Their expected performance index would be -34.075

```python
print(f"Coeffficient or Slope is {model.coef_}")
```

```python
x
```

KEY FINDING

- Hours studied has a significant positive impact on performance (Î² = 2.852)
- Previous scores are a strong predictor of future performance (Î² = 1.019)
- Sleep hours have a small positive effect on performance (Î² = 0.477)
- Simple question paper practiced has a minimal positive effect on performance (Î² = 0.188).
- Extracurricular activities have a moderate positive effect on performance (Î² = 0.665)

These findings suggest that: 

1. Encouraging students to study regularly can improve academic outcomes. 
2. Prior academic performance is a reliable indicator of future success. 
3. Extracurricular activities can enhance student engagement and performance. 
4. Adequate sleep hours contribute to better academic performance. 
5. Practicing simple question papers can have a slight positive impact.

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
print (f"mean square error {mse}")
```

```python
# Calculating the error metrices

def eval_metrices (actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    print ("Model testing Performance: ")
    print ("----------------------------")
    print (f"MAE \t\t: {mae}")
    print (f"r2_score \t: {R2_score}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
```

```python
eval_metrices (y_test, y_pred)
```

Interpretation:

- RÂ² (0.988): The model explains approximately 98.8% of the variance in student performance
- MSE (4.106): The average squared difference between predicted and actual values is relatively low
- RMSE (2.026): The square root of MSE, indicating a low average error
- MAE (1.623): The average absolute difference between predicted and actual values is relatively low.
